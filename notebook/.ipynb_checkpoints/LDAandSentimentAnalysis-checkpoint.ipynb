```python
import gensim
import gensim.corpora as corpora
from gensim.corpora import Dictionary
from gensim.models.coherencemodel import CoherenceModel
from gensim.models.ldamodel import LdaModel
from pprint import pprint
import spacy
import pickle
import re 
import pyLDAvis
import pyLDAvis.gensim_models
import matplotlib.pyplot as plt 
import pandas as pd
from nltk.corpus import wordnet as wn
import nltk
import itertools
import numpy as np
from textblob import TextBlob
from spacy.lang.en import English
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')

from nltk.tokenize import word_tokenize
import nltk
import seaborn as sns
import matplotlib
nltk.download('punkt')
```

```python
def read_data(file_name):
    return pd.read_csv(file_name)
```

```python
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma


from nltk.stem.wordnet import WordNetLemmatizer


def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

```python
def join_bigram(l):
    return " ".join([i.split()[0] for i in l])
```

```python
def evaluate_graph(dictionary, corpus, texts, limit):
    """
    Function to display num_topics - LDA graph using c_v coherence
    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    limit : topic limit
    Returns:
    -------
    lm_list : List of LDA topic models
    c_v : Coherence values corresponding to the LDA model with respective number of topics
    """
    c_v = []
    lm_list = []
    for num_topics in range(1, limit):
        lm = gensim.models.ldamulticore.LdaMulticore(corpus=corpus, num_topics=num_topics, id2word=dictionary,
                                                     iterations=500)
        lm_list.append(lm)
        cm = gensim.models.ldamodel.CoherenceModel(model=lm, texts=texts, dictionary=dictionary, coherence='c_v')
        c_v.append(cm.get_coherence())

    # Show graph
    x = range(1, limit)
    plt.plot(x, c_v)
    plt.xlabel("num_topics")
    plt.ylabel("Coherence score")
    plt.legend(("c_v"), loc='best')
    plt.show()

    return lm_list, c_v
```

```python
data_frame = read_data("Processed_reviews.csv")

parser = English()
en_stop = set(nltk.corpus.stopwords.words('english'))


# we apply a filter to get the most reviewed product
data_frame = data_frame.loc[data_frame['ProductId'] == "B003B3OOPA"]
reviews = data_frame.clean_text.values
text_data = []
```

```python
for r in reviews:
    tokens = prepare_text_for_lda(r)
    bigram = list(nltk.bigrams(tokens))
    tokens = []
    for i in bigram:
        tokens.append((''.join([w + ' ' for w in i])).strip())
    text_data.append(tokens)
    
wholetext = list(itertools.chain.from_iterable(text_data))
text = nltk.Text(wholetext)
# Calculate Frequency distribution
freq = nltk.FreqDist(text)
metadb = data_frame.shape[0] * 0.4

lenfreq = freq.most_common(20)
freqword = []
for i in lenfreq:
    if i[1] > metadb:
        freqword.append(i[0])

freq.plot(10)
```

```python
dictionary = corpora.Dictionary(text_data)
dictionary.filter_extremes(no_below=10, no_above=0.5)
corpus = [dictionary.doc2bow(text) for text in text_data]


# Finding out the optimal number of topics
np.random.seed(50)
lmlist, c_v = evaluate_graph(dictionary=dictionary, corpus=corpus, texts=text_data, limit=10)
max_value = max(c_v)
max_index = c_v.index(max_value)
NUM_TOPICS = max_index + 1

ldamodel = gensim.models.ldamulticore.LdaMulticore(corpus, num_topics=NUM_TOPICS, id2word=dictionary,
                                                       iterations=500)
topics = ldamodel.print_topics(num_words=6)
x = ldamodel.show_topics(num_topics=NUM_TOPICS, num_words=15, formatted=False)
topics_words = [(tp[0], [wd[0] for wd in tp[1]]) for tp in x]
```

```python
# Compute Coherence Score using c_v
r_list = []
prob_list = []
topic_list = []
for r in text_data:
    bow = dictionary.doc2bow(r)
    t = ldamodel.get_document_topics(bow)
    maxval = list(max(t, key=lambda i: i[1]))
    minval = list(min(t, key=lambda i: i[1]))
    print(t)
    print(maxval)
    print(minval)
    if (maxval[1] == minval[1]):
        maxval[0] = 0
    else:
        maxval[0] = maxval[0] + 1
    r_list.append(r)
    prob_list.append(maxval[1])
    topic_list.append(maxval[0])

data_frame_final = pd.DataFrame(data={'review': r_list, 'probability': prob_list, 'topic_no': topic_list})

for current_topic in data_frame_final.topic_no.unique():

    text_reviews = [join_bigram(i) for i in data_frame_final.loc[data_frame_final['topic_no'] == current_topic].review]
    sentiment_scores = list()
    for current_topic_reviews in text_reviews:
        if (current_topic_reviews != ''):
            line = TextBlob(current_topic_reviews)
            sentiment_scores.append(line.sentiment.polarity)
    print(np.mean(sentiment_scores))
    print("Current Topic = {}".format(current_topic))
```

```python
def reviews_sentiment():
    df = data_frame.dropna()

    stop = en_stop

    for index, row in df.iterrows():
        word_tokens = word_tokenize(row.clean_text)

        filtered_sentence = [w for w in word_tokens if not w in stop]

        filtered_sentence = []

        for w in word_tokens:
            if w not in stop:
                filtered_sentence.append(w)
                
        df.at[index, 'clean_text'] = " ".join(filtered_sentence)

    sentiment_scores = list()
    i = 0
    for sentence in df.clean_text:
        line = TextBlob(sentence)
        sentiment_scores.append(line.sentiment.polarity)

    df['polarity'] = sentiment_scores
    normalized_polarity = 2*(df['polarity'] - df['polarity'].min()) / (df['polarity'].max() - df['polarity'].min())-1
    normalized_score = 2*(df['Score'] - df['Score'].min()) / (df['Score'].max() - df['Score'].min())-1
    sns.distplot(normalized_polarity)
    sns.distplot(normalized_score)
    plt.show()

    print(np.corrcoef(df.Score, df.polarity))
    matplotlib.style.use('ggplot')

    plt.scatter(df.Score, df.polarity)
    plt.show()
```
